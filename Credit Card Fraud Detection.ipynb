# Data Analysis

## Time

```python
timedf = df[["Time", "Class"]]
timedf.loc[(timedf["Time"] <= 30000) & (timedf["Time"] > 0), "Time"] = 1
timedf.loc[(timedf["Time"] <= 60000) & (timedf["Time"] > 30000), "Time"] = 2
timedf.loc[(timedf["Time"] <= 90000) & (timedf["Time"] > 60000), "Time"] = 3
timedf.loc[(timedf["Time"] <= 120000) & (timedf["Time"] > 90000), "Time"] = 4
timedf.loc[(timedf["Time"] <= 150000) & (timedf["Time"] > 120000), "Time"] = 5
timedf.loc[(timedf["Time"] <= 180000) & (timedf["Time"] > 150000), "Time"] = 6
```

```python
time = timedf.groupby(["Time", "Class"])
timedf = time.value_counts().reset_index()
timetotal = timedf.groupby("Time")["count"].transform("sum")
timedf["Percentage"] = timedf["count"].div(timetotal).mul(100)
timedfyes = (timedf["Class"] == 1)
timedf[timedfyes].sort_values(by = "Percentage", ascending = False)
```

## General Trend

```python
nofraud = (df["Class"] == 0)
nofrauddf = df[nofraud]
nofrauddf.describe()
```

```python
fraud = (df["Class"] == 1)
frauddf = df[fraud]
frauddf.describe()
```

```python
lgrg = LogisticRegression()
```

```python
outcome = df.Class
predictors = df.drop("Class", axis = 1)

xtrain, xtest, ytrain, ytest = train_test_split(predictors, outcome, train_size = 0.8, random_state = 1)

lgrg.fit(xtrain, ytrain)
y_preds = lgrg.predict(xtest)
accuracy = accuracy_score(ytest, y_preds)
accuracy
```
